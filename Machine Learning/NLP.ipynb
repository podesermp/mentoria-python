# Com dataset

```python
abstract = pd.read_csv('./data/abstracts.csv')
abstract.head()
```

```python
abstract['tokenized'] = abstract['Abstract'].map(lambda t: word_tokenize(t))
```

```python
abstract['tokenized'] = abstract['tokenized'].map(lambda t: [word.lower().strip() for word in t if word.isalpha()])
```

```python
abstract['tokenized'] = abstract['tokenized'].map(lambda t: [wordnet_lemmatizer.lemmatize(word) for word in t if word not in stopwords_en])
```

## Usando o gensim

```python
from gensim.corpora.dictionary import Dictionary

dictionary = Dictionary(abstract['tokenized'])
```

```python
firm_id = dictionary.token2id.get('firm')
print(dictionary.get(firm_id))
```

```python
corpus = [dictionary.doc2bow(abstract) for abstract in abstract['tokenized']]
```

```python
print(corpus[0])
print(abstract['tokenized'][0])
```

```python
print(corpus[10][:10])
```

```python
bow_doc = sorted(corpus[10], key=lambda w: w[1], reverse=True)

for word_id, word_count in bow_doc[:10]:
    print(dictionary.get(word_id), word_count)
```

```python
from gensim.models.tfidfmodel import TfidfModel

tfidf = TfidfModel(corpus)

tfidf_weights = tfidf[corpus[10]]

print(tfidf_weights[:5])
```

```python
sorted_tfidf_weights = sorted(tfidf_weights, key=lambda w:w[1], reverse=True)

for term_id, weight in sorted_tfidf_weights[:10]:
    print(dictionary.get(term_id), weight)
```

```python
tfidf_corpus = tfidf[corpus]
```

```python
from gensim.models.lsimodel import LsiModel

lsi = LsiModel(tfidf_corpus, id2word=dictionary, num_topics=100)

lsi.show_topics(num_topics=10)
```

```python
lsi_corpus = lsi[tfidf_corpus]
```

```python
from gensim.similarities import MatrixSimilarity

document_topic_matrix = MatrixSimilarity(lsi_corpus)
document_topic_matrix = document_topic_matrix.index
```

```python
document_topic_matrix
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10)

kmeans.fit(document_topic_matrix)
```

```python
abstract['cluster'] = kmeans.labels_
```

```python
# Preprocessing
abstract['title_tok'] = abstract['Title'].map(lambda t: word_tokenize(t))
abstract['title_tok'] = abstract['title_tok'].map(lambda t: [word.lower().strip() for word in t if word.isalpha()])
abstract['title_tok'] = abstract['title_tok'].map(lambda t: [wordnet_lemmatizer.lemmatize(word) for word in t if word not in stopwords_en])
```

```python
# Collectiong

Cluster = 2

cluster_titles = []
for x in abstract[abstract['cluster'] == Cluster]['title_tok']:
    cluster_titles.extend(x)
```

```python
# Transfortm into tf_idf format
titles_tfidf = tfidf[dictionary.doc2bow(cluster_titles)]
```

```python
# Sort the weights from highest to lowest: sorted_tfidf_weights
titles_tfidf = sorted(titles_tfidf, key=lambda w: w[1], reverse=True)

# Print the top 5 weighted words
for term_id, weight in titles_tfidf[:20]:
    print(dictionary.get(term_id), weight)
```
