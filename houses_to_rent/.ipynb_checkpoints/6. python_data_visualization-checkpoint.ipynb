# 2. Visualização de Dados


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Carregando os dados na variável *data* e mudando os nomes das colunas com o símbolo do Real (R$) para facilitar o acesso aos valores dessas colunas.

```python
data = pd.read_csv("./houses_to_rent_v2.csv")

# mudando o nome das colunas para facilitar o acesso aos valores
data.rename(
    columns={
        "hoa (R$)": "hoa",
        "rent amount (R$)": "rent amount",
        "property tax (R$)": "property tax",
        "fire insurance (R$)": "fire insurance",
        "total (R$)": "total"
    },
    inplace=True
)
```

No primeiro contato com o *DataFrame* é possível notar a coluna *city* com valores categóricos. Note que é possível coletar estatísticas com relação aos alugueis por cidade.

```python
# mostrando as 10 primeiras linhas do DataFrame
data.head(10)
```

A função describe mostra que a despesa média é de quase 5000 com desvio padrão de aproximadamente 1600. Essa é uma descrição geral, para buscar descrições específicas para cada cidade é necessário visualizar o *DataFrame* por cidade.

```python
data.describe()
```

```python
fig, ax = plt.subplots(1, figsize=(12, 8))
sns.boxplot(x="city", y="total", data=data, ax=ax)
sns.stripplot(x="city", y="total", data=data, ax=ax)
plt.show()
```

O plot acima é um *boxplot* com a dispersão dos dados em torno da coluna *total*. Note que há alguns pontos que recaem em um intervalo fora daquele relativo a maior parte dos dados. Esse tipo de dado é denominado *outlier*. Um maneira simples de dectar *outlier* é utilizar medidas de dispersão como o **desvio padrão** e o **intervalo interquartil**.

* Desvio padrão:

    Defina $\mu$ a média de uma coluna:

$$\sigma = \frac{\sqrt{\sum_i(x_i - \mu)^2}}{n-1}.$$

* Intervalo interquartil (*interquartile range*):
    - $Q1$: valor aos 25% dos valores ordenados;
    - $Q2$: valor aos 50% dos valores ordenados (mediana);
    - $Q3$: valor aos 75% dos valores ordenados.
    
    O intervalo interquartil é calculado utilizando o primeiro e o terceiro quartil:
    
    $$IRQ = Q3 - Q1.$$

* Detectando os *outliers* com o Desvio padrão (https://pt.wikipedia.org/wiki/Desvio_padr%C3%A3o):
    Estatísticamente, 95% dos dados estão em até 2 desvios padrões de distância da média.

```python
mean_total = data["total"].mean() # cálculo da média
std_total = data["total"].std() # cálculo do desvio padrão
data_inlier_std = data[(data["total"] >= (mean_total - std_total * 2)) & (data["total"] <= (mean_total + std_total * 2))]

print(f"Percentage of droped data: {(1 - data_inlier_std.shape[0]/data.shape[0]) * 100:.2f}")

fig, ax = plt.subplots(1, figsize=(12, 8))
sns.boxplot(x="city", y="total", data=data_inlier_std, ax=ax)
sns.stripplot(x="city", y="total", data=data_inlier_std, ax=ax)
plt.show()
```

* Detectando os *outliers* via intervalo interquartil (https://pt.wikipedia.org/wiki/Diagrama_de_caixa). O limite inferior e o limite superior são definidos, respectivamente, como:

$$LI = Q1 - c \cdot IRQ$$
$$LS = Q3 + c \cdot IRQ$$

* Comumente é utilizado $c = 1.5$ pois o valor é capaz de capturar mais de 99% dos dados sob uma curva normal em torno da mediana.

```python
Q1, Q3 = data["total"].quantile(0.25), data["total"].quantile(0.75) # cálculo do primeiro e terceiro quartil
irq_total = Q3 - Q1
data_inlier_irq = data[(data["total"] >= (Q1 - irq_total * 1.5)) & (data["total"] <= (Q3 + irq_total * 1.5))]

print(f"Percentage of droped data: {(1 - data_inlier_irq.shape[0]/data.shape[0]) * 100:.2f}")

fig, ax = plt.subplots(1, figsize=(12, 8))
sns.boxplot(x="city", y="total", data=data_inlier_irq, ax=ax)
sns.stripplot(x="city", y="total", data=data_inlier_irq, ax=ax)
plt.show()
```

Uma outra forma de buscar por *outliers* é utilizando a própria função de *boxplot* do pandas

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 8))
data.boxplot(["total"], ax=ax1)
data_inlier_std.boxplot(["total"], ax=ax2)
data_inlier_irq.boxplot(["total"], ax=ax3)
```

Uma vez removido os *outliers* a visualização dos dados fica mais concreta. Para se ter uma noção de quantas casas estão disponíveis por cidade pode-se utilizar a função *hist*.

```python
data_inlier_irq["city"].hist()
```

Pode-se ainda visualizar a distribuição das despesas totais por cidade. Abaixo, note que a densidade dos preços aumentam por volta de 3000. Ainda, as casas com despesas menos custosas estão localizadas em Porto Alegre e Campinas.

```python
sns.FacetGrid(data_inlier_irq, hue="city", height=6).map(sns.kdeplot, "total").add_legend()
```
