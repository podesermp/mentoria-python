## Pre-processamento

```python
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
from dictionary import morphy_tag
def preprocess_and_tokenze(text):
    
    ## Pre-processamento
    text = text.lower() # tudo minúsculo
    text = decontracted(text) # remoção das palavras contraidas
    text = re.sub(r'[^\w\s]', ' ', text) # remoção da pontuação
    text = text.replace('\n', ' ') # remove \n
    text = re.sub(r"[0-9]+", '', text) # remover números
    #text = " ".join(word.strip(string.punctuation) for word in text.split()) # remove o espaçamento no inicio e no fim 
    
    nlp = spacy.load('en_core_web_sm')
    doc = nlp(text)
    
    ## tokenização
#     text = word_tokenize(text)
    
    ## lemmatização/stemmer
#     text_tag = nltk.pos_tag(text)
#     dic = morphy_tag
#     lemmatizer = WordNetLemmatizer()
#     text = [lemmatizer.lemmatize(token, dic[pos]) for token, pos in text_tag]
    
#     stemmer = PorterStemmer()
#     text = [stemmer.stem(token) for token in text]
        
    ## remoção de stopwords
    stop = set(stopwords.words('english'))
    text = [phrase.lemma_ for phrase in doc if phrase not in stop]
    
    return text
```

```python
def preprocessing(text):
    text = text.lower() # tudo minúsculo
    text = decontracted(text) # remoção das palavras contraidas
    
    text = re.sub(r'[^\w\s]', ' ', text) # remoção da pontuação
    text = text.replace('\n', ' ') # remove \n
    text = re.sub(r"  ", " ", text) # remove espaços duplos
    
    text = " ".join(word.strip(string.punctuation) for word in text.split()) # remove o espaçamento no inicio e no fim 
    return text
```

```python
phrases = [preprocessing(word) for word in data['comment_text']]
```
