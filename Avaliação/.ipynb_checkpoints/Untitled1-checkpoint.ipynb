```python
import numpy as np
import pandas as pd
import re, string, spacy

import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, TweetTokenizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
```

## Carregamento do dataset oficial

```python
data = pd.read_csv('Data/dados_tratados/data_binary.csv')
```

```python
data = data.drop(115606) # é so um teste
```

## Funções de pré-processamento

```python
stemmer = PorterStemmer()
def preprocess_and_tokenze(text):
    
    ## Pre-processamento
    text = text.lower() # tudo minúsculo
    text = decontracted(text) # remoção das palavras contraidas
    text = re.sub(r'[^\w\s]', ' ', text) # remoção da pontuação
    text = text.replace('\n', ' ') # remove \n
    text = re.sub(r"[0-9]+", '', text) # remover números

    ## tokenização
    text = word_tokenize(text)
    
    ## remoção de stopwords e stemming
    stop = set(stopwords.words('english'))
    text = [stemmer.stem(phrase) for phrase in text if phrase not in stop]
    
    return text
```

```python
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

## Pré processamento

```python
texts = list()
for i,phrase in enumerate(data['text']):
    texts.append(" ".join(preprocess_and_tokenze(phrase)))
    print(f"{i+1}/{data.shape[0]}", end='\r')
```

## Salvando dataset pré-processado, tokenizado e stemmezado

```python
## isso é só pra salvar o dataset preprocessado, tokenizado e stemmezado
obj_text = {
    'text': texts,
    'toxic': data['toxic']
}
data_preprocess_tokenzed_stemmed = pd.DataFrame(obj_text, columns=['text', 'toxic'])
data_preprocess_tokenzed_stemmed.to_json('./Data/dados_tratados/data_preprocess_tokenzed_stemmed.json')
```
