## Pre-processamento

> Os dados foram pré-processados da seguinte forma:
* Todos os textos se tornaram minúsculos.
* Descontração de palavra contraídas (ex.: i'm, can't).
* Remoção de pontuação.
* Remoção de quebra de linhas (\n).
* Remoção de números.
* Tokenização(transformar a frase em tokens) dos textos.
* Stemização dos tokens.
* Remoção dos stopwords.

```python
stemmer = PorterStemmer()
def preprocess_and_tokenze(text):
    
    ## Pre-processamento textual
    text = text.lower() # tudo minúsculo
    text = decontracted(text) # descontração das palavras contraidas
    text = re.sub(r'[^\w\s]', ' ', text) # remoção da pontuação
    text = text.replace('\n', ' ') # remove \n
    text = re.sub(r"[0-9]+", '', text) # remover números

    ## tokenização
    text = word_tokenize(text)
    
    ## remoção de stopwords e stemming
    stop = set(stopwords.words('english'))
    text = [stemmer.stem(phrase) for phrase in text if phrase not in stop]
    
    return text
```

```python
def decontracted(phrase):
    # specific
    phrase = re.sub(r"won't", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)

    # general
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    return phrase
```

```python
'''
Execução da função de pré-processamento. 
Os textos depois de serem pré-processados são colocados na list 'texts' que no fim do 'for' tem os textos da mesma forma que os do dataset 'data_adjusted'.
'''
texts = list()
for phrase in tqdm(data['text']):
    texts.append(" ".join(preprocess_and_tokenze(phrase)))
```

```python
data_adjusted = pd.read_json('/content/gdrive/MyDrive/ColabNotebooks/ML/Avaliação/data_preprocess_tokenzed_stemmed.json')
```

## Vetorização
> Com as frases do dataset pré-processadas, nessa etapa cada frase é transformada em um vetor de números que irá ser usada para colocar no modelo de classificação. Para fazer essa vetorização foi usado o TfidfVectorizer que  converte uma coleção de documentos brutos em uma matriz de recursos do TF-IDF.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data_adjusted['text'])
y = data_adjusted['toxic'].to_numpy()
```

## Classificadores

```python
from sklearn.metrics import confusion_matrix, accuracy_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

### Execução

```python
from sklearn.model_selection import StratifiedKFold

kfold = StratifiedKFold(n_splits=3, random_state=1, shuffle=True)

lr_scores = {
    'accuracy': list(),
    'roc_auc': list(),
    'confusion_matrix': list()
}
random_scores = {
    'accuracy': list(),
    'roc_auc': list(),
    'confusion_matrix': list()
}

for train_idx, test_idx in tqdm(kfold.split(X,y)):
  X_train = X[train_idx]
  X_test = X[test_idx]
  y_train = y[train_idx]
  y_test = y[test_idx]

  ## LogisticRegression()
  clf_logistic = LogisticRegression()
  clf_logistic.fit(X_train,y_train)
  y_hat = clf_logistic.predict(X_test)
  lr_scores['accuracy'].append(accuracy_score(y_test, y_hat))
  lr_scores['roc_auc'].append(roc_auc_score(y_test, y_hat))
  lr_scores['confusion_matrix'].append(confusion_matrix(y_test, y_hat))

  ## RandomForestClassifier()
  clf_random = RandomForestClassifier()
  clf_random.fit(X_train,y_train)
  y_hat = clf_random.predict(X_test)
  random_scores['accuracy'].append(accuracy_score(y_test, y_hat))
  random_scores['roc_auc'].append(roc_auc_score(y_test, y_hat))
  random_scores['confusion_matrix'].append(confusion_matrix(y_test, y_hat))

save_results(lr=lr_scores, random=random_scores)
```

### Salvar os resultados

```python
def save_results(lr, random):
  lr = pd.DataFrame(lr)
  lr.to_json('/content/gdrive/MyDrive/ColabNotebooks/ML/Avaliação/results/lr_scores.json') 

  random = pd.DataFrame(random)
  random.to_json('/content/gdrive/MyDrive/ColabNotebooks/ML/Avaliação/results/random_scores.json') 
```

### Visualização das métricas

```python
data_lr_scores = pd.read_json('/content/gdrive/MyDrive/ColabNotebooks/ML/Avaliação/results/lr_scores.json')
data_random_scores = pd.read_json('/content/gdrive/MyDrive/ColabNotebooks/ML/Avaliação/results/random_scores.json')
```

#### LogisticRegression

```python
lr_mc = np.mean(data_lr_scores['confusion_matrix'].to_list(), axis=0)

print('Dados do LogisticRegression')
print(f"Acurácia: {np.mean(data_lr_scores['accuracy'])*100:.2f}%")
print(f"Roc_auc: {np.mean(data_lr_scores['roc_auc']):.2f}\n")
print(f"matriz de confusão: \n{np.mean(data_lr_scores['confusion_matrix'].to_list(), axis=0)}\n")

plt.title('LogisticRegression')
sns.heatmap(lr_mc/np.sum(lr_mc), annot=True, fmt='.2%', cmap='Blues')
```

```python
random_mc = np.mean(data_random_scores['confusion_matrix'].to_list(), axis=0)

print('Dados do RandomForestClassifier')
print(f"Acurácia: {np.mean(data_random_scores['accuracy'])*100:.2f}%")
print(f"Roc_auc: {np.mean(data_random_scores['roc_auc']):.2f}\n")
print(f"matriz de confusão: \n{np.mean(data_random_scores['confusion_matrix'].to_list(), axis=0)}\n")

plt.title('RandomForestClassifier')
sns.heatmap(random_mc/np.sum(random_mc), annot=True, fmt='.2%', cmap='Blues')
```
