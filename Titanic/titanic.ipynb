```python
def pclass_name(x):
    if x == 1:
        x = '1st Class'
    if x == 2:
        x = '2st Class'
    if x == 3:
        x = '3st Class'
    return x
```

```python
df3['Pclass'] = df3['Pclass'].apply(lambda x: pclass_name(x))
# df3['Pclass'] = df3['Pclass'].apply(pclass_name)
df3
```
