# Fly dalays

```python
flight_data = pd.read_csv('/home/podesermp/Área de Trabalho/Estágio - Lead/Bolsa Python/mentoria/curso kaggle/flight_delays.csv', index_col='Month')
flight_data
```

# Seguros

```python
insurance_data = pd.read_csv('/home/podesermp/Área de Trabalho/Estágio - Lead/Bolsa Python/mentoria/curso kaggle/insurance.csv')
insurance_data.head()
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(x=insurance_data['bmi'], y = insurance_data['charges'])
```

```python
plt.figure(figsize=(14,6))
sns.regplot(x=insurance_data['bmi'], y = insurance_data['charges']) # com linha de regressão para evidenciar que há uma correlação positiva entre as pessoas com IMC alto e encargos de seguro
```

```python
# com 3 variáveis
plt.figure(figsize=(14,6))
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

```python
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

```python
plt.figure(figsize=(10,5))
sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['charges'])
```

# Flores - Não entendi direito esse tipo de gráfico

```python
iris_data = pd.read_csv('/home/podesermp/Área de Trabalho/Estágio - Lead/Bolsa Python/mentoria/curso kaggle/iris.csv', index_col='Id')
iris_data.head()
```

```python
sns.distplot(a=iris_data['Petal Length (cm)'], kde=True)
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)  # shade é para colocar a sombra na onda
```

```python
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
```

```python
iris_set_data = iris_data[iris_data['Species'] == 'Iris-setosa']
iris_ver_data = iris_data[iris_data['Species'] == 'Iris-versicolor']
iris_vir_data = iris_data[iris_data['Species'] == 'Iris-virginica']
```

```python
sns.distplot(a=iris_set_data['Petal Length (cm)'], label='Iris-setosa', kde=False)
sns.distplot(a=iris_ver_data['Petal Length (cm)'], label="Iris-versicolor", kde=False)
sns.distplot(a=iris_vir_data['Petal Length (cm)'], label="Iris-virginica", kde=False)

plt.title("Histogram of Petal Lengths, by Species")

plt.legend()
```

```python
sns.kdeplot(data=iris_set_data['Petal Length (cm)'], label="Iris-setosa", shade=True)
sns.kdeplot(data=iris_ver_data['Petal Length (cm)'], label="Iris-versicolor", shade=True)
sns.kdeplot(data=iris_vir_data['Petal Length (cm)'], label="Iris-virginica", shade=True)

plt.title("Distribution of Petal Lengths, by Species")
```
